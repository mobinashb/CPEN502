```python
with open('a-trainingErrors.txt', 'r') as f:
    a_errors = [float(line.strip()) for line in f]
with open('b-trainingErrors.txt', 'r') as f:
    b_errors = [float(line.strip()) for line in f]
with open('c-trainingErrors.txt', 'r') as f:
    c_errors = [float(line.strip()) for line in f]
```

# Part A: binary & momentum = 0

```python
from matplotlib import pyplot as plt
```

```python
colors = ['#0099ff', '#ff3300', '#00cc00']
```

```python
plt.figure(figsize=(10,7))
line = plt.plot(a_errors, color=colors[0])
plt.annotate("epoch: {}".format(len(a_errors)-1), xy =(len(a_errors)-1, a_errors[-1]),
                xytext =(len(a_errors) - 270, a_errors[-1] + 0.05), 
                arrowprops = dict(facecolor =colors[0], shrink = 0.05),)

plt.title("Part A")
plt.xlabel("Epochs")
plt.ylabel("Total Error")
  
plt.show()
```
